#### 1. LSTM

- RNN에서 기울기 소멸 문제를 해결하기 위해서 개발

- 망각 게이트, 입력 게이트, 출력 게이트를 사용하여 기울기 소멸 문제 해결

##### 1-1) LSTM 순전파

##### 망각 게이트: 시그모이드로 계산하여 1인 것만 통과

- ${f_t = \sigma(w_f[h_{t-1}, x_t])}$

- ${c_t = f_t \times c_{t-1}}$

##### 입력 게이트: 시그모이드와 하이퍼볼릭 탄젠트 함수로 보존량 결정
