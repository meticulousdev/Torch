#### Pytorch

- Input: $(N, C_{in}, H_{in}, W_{in})$ or $(C_{in}, H_{in}, W_{in})$

- Output: $(N, C_{out}, H_{out}, W_{out})$ or $(C_{out}, H_{out}, W_{out})$

[Conv2d](https://pytorch.org/docs/stable/generated/torch.nn.Conv2d.html)

torch.nn.Conv2d(in_channels, out_channels, kernel_size, stride=1, padding=0, dilation=1, groups=1, bias=True, padding_mode='zeros', device=None, dtype=None)

- $H_{out}=[{{H_{in} + 2 \times padding[0] - dilation[0] \times (kernel\_size[0] - 1) - 1}\over{stride[0]}} + 1]$

- $W_{out}=[{{W_{in} + 2 \times padding[1] - dilation[1] \times (kernel\_size[1] - 1) - 1}\over{stride[1]}} + 1]$

[MaxPool2d](https://pytorch.org/docs/stable/generated/torch.nn.MaxPool2d.html)

torch.nn.MaxPool2d(kernel_size, stride=None, padding=0, dilation=1, return_indices=False, ceil_mode=False)

- $H_{out}=[{{H_{in} + 2 \times padding[0] - dilation[0] \times (kernel\_size[0] - 1) - 1}\over{stride[0]}} + 1]$

- $W_{out}=[{{W_{in} + 2 \times padding[1] - dilation[1] \times (kernel\_size[1] - 1) - 1}\over{stride[1]}} + 1]$

[AdaptiveAvgPool2d](https://pytorch.org/docs/stable/generated/torch.nn.AdaptiveAvgPool2d.html)

torch.nn.AdaptiveAvgPool2d(output_size)

[BatchNorm2d](https://pytorch.org/docs/stable/generated/torch.nn.BatchNorm2d.html)

torch.nn.BatchNorm2d(num_features, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True, device=None, dtype=None)

eps($\epsilon$) = ${10^{-5}}$ (default)

momentum = ${0.1}$ (default)

affine = ${True}$ (default)

- $y={{x - E[x]}\over{Var[x]+\epsilon}}\times{\gamma} + \beta$

[argmax](https://pytorch.org/docs/stable/generated/torch.argmax.html)

torch.argmax(input, dim, keepdim=False) → LongTensor

[view_as](https://pytorch.org/docs/stable/generated/torch.Tensor.view_as.html)

Tensor.view_as(other) → Tensor

[view](https://pytorch.org/docs/stable/generated/torch.Tensor.view.html#torch.Tensor.view)

Tensor.view(*shape) → Tensor

[squeeze](https://pytorch.org/docs/stable/generated/torch.squeeze.html)

torch.squeeze(input, dim=None, *, out=None) → Tensor

- Returns a tensor with all the dimensions of input of size 1 removed.

```python
import torch

x = torch.zeros(2, 1, 2, 1, 2)
print(x)
print(x.size())

y = torch.squeeze(x)
print(y)
print(y.size())
```

[chunk](https://pytorch.org/docs/stable/generated/torch.chunk.html)

torch.chunk(input, chunks, dim=0) → List of Tensors

- Attempts to split a tensor into the specified number of chunks. Each chunk is a view of the input tensor.
